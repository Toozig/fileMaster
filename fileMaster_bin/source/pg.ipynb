```python
import os


FILE_TYPE_DICT= {
    'Table': 'For table file, e.g. .csv, .tsv (bed files are table as well, and also vcf 🤯)',
    'Bed': 'Specific for bed files, have same functionality as table files... and even more 🤯',
    'PBM': 'For PBM expirement result',
    'Generic Source': 'Minimal information about the file. This won\'t give the full functionality of the file package. Don\'t be lazy 😤.',
}

def print_ascii_art():
    # ASCII art for cool header
    ascii_art = """
oooooooooooo  o8o  oooo            ooo        ooooo                        .                                .o     .oooo.     .oooo.   
`888'     `8  `"'  `888            `88.       .888'                      .o8                              .d88   .dP""Y88b  .dP""Y88b  
 888         oooo   888   .ooooo.   888b     d'888   .oooo.    .oooo.o .o888oo  .ooooo.  oooo d8b       .d'888         ]8P'       ]8P' 
 888oooo8    `888   888  d88' `88b  8 Y88. .P  888  `P  )88b  d88(  "8   888   d88' `88b `888""8P     .d'  888       <88b.      .d8P'  
 888    "     888   888  888ooo888  8  `888'   888   .oP"888  `"Y88b.    888   888ooo888  888         88ooo888oo      `88b.   .dP'     
 888          888   888  888    .o  8    Y     888  d8(  888  o.  )88b   888 . 888    .o  888              888   o.   .88P  .oP     .o 
o888o        o888o o888o `Y8bod8P' o8o        o888o `Y888""8o 8""888P'   "888" `Y8bod8P' d888b            o888o  `8bd88P'   8888888888 
                                                                                                                                                                                                                                                                                                                                                                                    
    """
    print(ascii_art)

def welcome_message():
    print("\n🌟 Welcome to FileMaster 432 - The -not-so-Ultimate-but-what-we-have-🤷‍♀️ Server File Organizer! 🌟")
    print("Our files are about to embark on a wild journey through the digital cosmos (e.g. DSI server and Dropbox).")
    # print("Just like the Hitchhiker's Guide to the Galaxy, but for your server files!")
    print("Hold tight, and let's get ready for some organized chaos! 🚀📂")


def get_username():
    username = input("\nEnter your username (or just hit ENTER to use your Linux username): ")
    return username.strip() if username else os.getlogin()  # Use Linux username

def check_and_print_file(file_path):
    if os.path.exists(file_path):
        print("\nFile exists! 😲 Amazeballs!")
    else:
        print("\nFile not found. 😠 Ugh, seriously?!")

def ask_for_file_type():
    print("\nPlease choose the file type:")
    for index, key in enumerate(FILE_TYPE_DICT.keys()):
        print(f"{index}. {key} - {FILE_TYPE_DICT[key]}")
        print("If you think the file is not any of the above, please choose 'Generic Source'.")
        print("Or you can just write a new file type 😇")
        print("(Choose generic source if you are not sure about the file type 🫠)")

    while True:
        try:
            file_type_index = int(input("Enter the number of the file type: "))
            if file_type_index in range(len(FILE_TYPE_DICT)):
                return list(FILE_TYPE_DICT.keys())[file_type_index]
            else:
                print("Invalid input. Please enter a valid number.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")



if __name__ == "__main__":
    print_ascii_art()
    welcome_message()
    username = get_username()

    while True:
        # Get file path from the user
        file_path = input(f"\nHey {username}! Enter the file path to check (Ctrl+C to exit): ")

        try:
            # Check and print the status
            check_and_print_file(file_path)
            
            # Ask for file type information
            file_type = ask_for_file_type()
            print(f"\nYou have classified the file as: {file_type}")
            break  # Break the loop if the file path is valid
        except KeyboardInterrupt:
            print("\nSo long, and thanks for all the files! Goodbye! 🐡🦈")
```

```python
import json
def write_ascii_art(string):
    with open('ascii_art.json', 'r') as f:
    
        asdcii_dict = json.load(f)
    letters = [asdcii_dict[letter.lower()].split('\n') for letter in string]
    for i in range(len(letters[0])):
        print(''.join([letter[i] for letter in letters]))

write_ascii_art('Roni')
```

```python
list_of_dicts = [
    {'id': 1, 'name': 'John'},
    {'id': 2, 'name': 'Jane'},
    {'id': 1, 'name': 'John'},  # Duplicate
    {'id': 3, 'name': 'Bob'},
]

def has_duplicates(lst):
    seen = set()
    for d in lst:
        # Convert each dictionary to a frozenset to make it hashable
        frozen_dict = frozenset(d.items())
        if frozen_dict in seen:
            return True
        seen.add(frozen_dict)
    return False

if has_duplicates(list_of_dicts):
    print("Duplicates found!")
else:
    print("No duplicates.")
```
